```python
df_temp['Date'] = df_temp['Year'].astype(str)+'-'+df_temp['Month'].astype(str)+'-'+df_temp['Day'].astype(str)
```

```python
df_temp = df_temp.rename(columns={"AvgTemperature": "AvgTemperature_F"})
```

```python
df_temp['AvgTemperature_C'] = (df_temp['AvgTemperature_F']-32)*5.0/9.0
```

```python
df_temp['AvgTemperature_C'] = df_temp['AvgTemperature_C'].round(2)
```

```python
df_temp.loc[(df_temp['Country'] == 'US'), 'Country'] = 'USA'
```

```python
df_temp.loc[(df_temp['Country'] == 'New Zealand'), 'Country'] = 'NZ'
```

```python
df_temp = df_temp.groupby(['Country', 'Date'], as_index=False, sort=False)['AvgTemperature_C'].mean()
```
